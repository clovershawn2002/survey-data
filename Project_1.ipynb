## NO.1 QUESTION; AT WHAT COMPANIES DO DEVELOPERS GET PAID THE MOST?

```python
import pandas as pd
# Group the data by the company column and calculate the average salary
average_salary_by_company = df.groupby('Industry')['CompTotal'].mean()
# Determine which company pays more than the others
highest_paid_company = average_salary_by_company.idxmax()
lowest_paid_company = average_salary_by_company.idxmin()
# Print the result
print(average_salary_by_company)
print(f"The company paying the highest average salary is: {highest_paid_company}")
print(f"The company paying the lowest average salary is: {lowest_paid_company}")
```

## NO.2 QUESTION; HOW MUCH DOES REMOTE WORKING MATTERS TO EMPLOYEES?

```python
import pandas as pd

# Count the occurrences of each unique value in the 'RemoteWork' column
importance_counts = df['RemoteWork'].value_counts()

# Sort the values by their counts
remote_importance = importance_counts.idxmax()

# Assuming you have a DataFrame named 'df' with a column 'RemoteWork'
column_data = df['RemoteWork']

# Create a new column 'RemoteWorkImportance' in the DataFrame
df['RemoteWorkImportance'] = ''

for i, value in enumerate(column_data):
    if value == "Hybrid (some remote, some in-person)":
        df.at[i, 'RemoteWorkImportance'] = "matters"
    elif value == "Remote":
        df.at[i, 'RemoteWorkImportance'] = "matters a lot"
    elif value == "In-person":
        df.at[i, 'RemoteWorkImportance'] = "does not matter"
    else:
        df.at[i, 'RemoteWorkImportance'] = "nan"



print(importance_counts)
print(f"How much does RemoteWorking matter to developers: {remote_importance}")
```

## NO.3 QUESTION; HOW  DOES CODING EXPERIENCE AFFECT THE LEVEL OF PAY?

```python
import pandas as pd
# Group the data by the year fo coding column and calculate the average salary
pay_by_code_experience = df.groupby('YearsCode')['CompTotal'].mean()
# Determine which coding experience has the highest pay
highest_paid_by_experience = pay_by_code_experience.idxmax()
# Print the result
print(pay_by_code_experience)
print(f"Does the years of experince affect your pay: {highest_paid_by_experience}")
```
